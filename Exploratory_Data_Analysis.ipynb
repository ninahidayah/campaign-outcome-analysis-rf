```python
import pandas as pd
```

```python
# read data
data = pd.read_csv('/content/drive/MyDrive/python_g.colab/Membuat Portofolio/bank-full.csv', sep=';')
```

```python
data.head()
```

```python
# rename column
data.rename(columns = { 'y' : 'deposit_status'}, inplace = True)
```

```python
data.head()
```

```python
# cek nilai unique from column age
data_age = data['age'].sort_values(ascending=True).unique()
data_age
```

```python
# create category age
def categorize_age(age):
    if age <= 25:
        return 'Young'
    elif 26 <= age <= 35:
        return 'Adult'
    elif 36 <= age <= 50:
        return 'Mature'
    elif 51 <= age <= 65:
        return 'Senior'
    else:
        return 'Retired'

data['category_age'] = data['age'].apply(categorize_age)
data.head()
```

Data Preparation

```python
# cek format tipe data
data.info()
```

```python
# cek missing value
data.isnull().sum()
```

```python
# cek data duplicat
data.duplicated().sum()
```

```python
prop_deposit = data.groupby(by = 'deposit_status', as_index = False).agg(total_campaign_outcomes = ('deposit_status', 'count'))
prop_deposit
```

```python
import matplotlib.pyplot as plt
```

```python
# show with visualisation
import plotly.express as px

# create pie chart
fig = px.pie(
    prop_deposit,
    values = 'total_campaign_outcomes',
    names = 'deposit_status',
    hole = 0.5,
    color_discrete_sequence = ['#750E21', '#9EC8B9']
)
# title position
fig.update_layout(
    width = 600,
    height = 600,
    showlegend = False,
    title_x=0.5,
    title = (
        "<b>Proporsi Respon Campaign Deposito</b>"
    )
)
# label position
fig.update_traces(
    textposition = 'outside',
    textinfo = 'percent+label',
    hovertemplate = '<b>%{label}</b><br>Jumlah: %{value}<br>Persentase: %{percent}'

)
fig.show()
```

View Previous outcome vs current Campaign Outcome

```python
poutcome_count = data.groupby(by = ['poutcome', 'deposit_status'], as_index = False ).size().rename(columns={'size': 'count_previous_outcomes'})
poutcome_count
```

```python
#total per poutcome
total_per_poutcome = poutcome_count.groupby('poutcome')['count_previous_outcomes'].transform('sum')

# persentase
poutcome_count['prop_previous_outcomes'] = round((poutcome_count['count_previous_outcomes'] / total_per_poutcome) * 100, 2)

poutcome_count
```

```python
fig = px.bar(poutcome_count,
             x='prop_previous_outcomes',
             y='poutcome',
             color='deposit_status',
             orientation='h',
             color_discrete_sequence=['#750E21', '#9EC8B9'],
             title='<b>Proporsi Respon Campaign Berdasarkan Previous Outcome</b>',
             barmode='stack',
             text='prop_previous_outcomes')

fig.update_traces(texttemplate='%{text:.1f}%', textposition='inside')

fig.update_layout(xaxis_title="Persentase (%)",
                  yaxis_title="Previous Outcome",
                  legend_title="Deposit Status",
                  bargap=0.3)

fig.show()
```

View job vs Campaign Outcome

```python
prop_deposit_job = data.groupby(by = ['deposit_status', 'job'], as_index=False).agg(total_per_deposit = ('deposit_status', 'count'))
prop_deposit_job
```

```python
# total nasabah per job (Yes+No)
total_job = data.groupby('job', as_index=False).agg(total_job=('deposit_status', 'count'))
total_job
```

```python
merged = prop_deposit_job.merge(total_job, on='job')
merged
```

```python
# merge two tabels
merged = prop_deposit_job.merge(total_job, on='job')

# persentase
merged['percentage'] = round((merged['total_per_deposit'] / merged['total_job']) * 100, 1)
```

```python
fig = px.bar(merged,
             x='job',
             y='percentage',
             color='deposit_status',
             color_discrete_sequence = ['#750E21', '#9EC8B9'],
             barmode='stack',
             title='<b>Proporsi Respon Campaign Deposito per Job</b>')
fig.show()
```
